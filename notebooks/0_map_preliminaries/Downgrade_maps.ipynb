```python
import matplotlib.pyplot as plt
import numpy as np
import healpy as hp
import polymv
import pylab as pl
import math
import sys
sys.path.insert(1, '../../scripts')
import CMBanom
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
Nside_out = 16
```

```python
dir_Planck = "../../data/Planck_CMB/"
out_dir = "../../data/real/"
names_Planck = ["COM_CMB_IQU-commander_2048_R3.00_full.fits", "COM_CMB_IQU-nilc_2048_R3.00_full.fits", "COM_CMB_IQU-sevem_2048_R3.01_full.fits", "COM_CMB_IQU-smica_2048_R3.00_full.fits"]
labels_Planck = ["commander", "nilc", "sevem", "smica"]
```

```python
#for n in range(len(names_Planck)):
for n in range(len(labels_Planck)):
    cmb_map_hr = hp.read_map(dir_Planck+names_Planck[n])*1e6 #mK
    m = CMBanom.downgrade_map(cmb_map_hr, Nside_out)
    hp.mollview(m, title=labels_Planck[n])
    outf = out_dir+"map_"+labels_Planck[n]+"_nside_"+str(Nside_out)+".fits"
    hp.write_map(outf, m, overwrite=True, dtype=np.float64)
```
